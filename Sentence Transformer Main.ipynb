```python
from sentence_transformers import SentenceTransformer, util
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
import pandas as pd
data = pd.read_csv('samples.csv') 
```

```python
data.head()
```

```python
 sentences = data['sentence']
```

```python
 sentences
```

```python
sents_1 = sentences[0:31] #series
```

```python
list_ser1 = sents_1.tolist()
```

```python
sents_2 = sentences[32:64]
```

```python
list_ser2 = sents_2.tolist()
```

```python
sent1 = data.sentence.tolist()
```

```python
sent1
```

```python
#Encode all sentences
embeddings1 = model.encode(list_ser1)
embeddings2 = model.encode(list_ser2)
```

```python
#Compute cosine similarity between all pairs
cos_sim = util.cos_sim(embeddings1, embeddings2)
```

```python
#Add all pairs to a list with their cosine similarity score
all_sentence_combinations = []
for i in range(len(cos_sim)-1):
    for j in range(i+1, len(cos_sim)):
        all_sentence_combinations.append([cos_sim[i][j], i, j])
```

```python
all_sentence_combinations
```

```python
#Sort list by the highest cosine similarity score
all_sentence_combinations = sorted(all_sentence_combinations, key=lambda x: x[0], reverse=True)
```

```python
print("Top-5 most similar pairs:")
x=[]
for score, i, j in all_sentence_combinations[0:5]:
    x.append([sentences[i], sentences[j], cos_sim[i][j]])
    print("{} \t {} \t {:.4f}".format(sentences[i], sentences[j], cos_sim[i][j]))
```

```python
x
```

```python
x=[]
for score, i, j in all_sentence_combinations[:]:
    x.append([sentences[i], sentences[j], cos_sim[i][j]])
    print("{} \t {} \t {:.4f}".format(sentences[i], sentences[j], cos_sim[i][j]))
```

```python
# Calling DataFrame constructor on list
df = pd.DataFrame(x)
df
```

```python
df.to_csv("all_sentence_combinations.csv")
```
